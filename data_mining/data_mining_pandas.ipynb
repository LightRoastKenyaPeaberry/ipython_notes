## 数据离散化

### one hot vector / 哑变量
### 连续属性的离散化就是将连续属性的值域上划分若干个离散区间， 最后用不同的符号或整数，值代表落在每个子区间的属性值。


### 如何实现数据的离散化： 
+ 分组
    + 自动分组 sr=pd.qcut(data, q)
    + 自定义分组 sr=pd.cut(data, [ ])
+ 将分组好的结果转换成one-hot vector
    + pd.get_dummies(sr, prefix=)

```python
# 1. 准备数据
data = pd.Series([165,174,160,180,159,163,192,184], index=['No1:165', 'No2:174','No3:160', 'No4:180', 'No5:159', 'No6:163', 'No7:192', 'No8:184']) 
```

```python
data
```

```python
# 2. 分组
# 自动分组
sr = pd.qcut(data, q=3)
```

```python
sr
```

```python
# 看每组有几条记录
sr.value_counts()
```

```python
# 自定义分组
sr_prime = pd.cut(data, [150,165,180,195])
```

```python
sr_prime
```

```python
sr_prime.value_counts()
```

```python
# 3. 转换成one-hot
pd.get_dummies(sr, prefix='height')
```

```python
pd.get_dummies(sr_prime, prefix='height')

# pd.get_dummies()
```
